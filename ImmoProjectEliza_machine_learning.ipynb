# Compare models

```python
# load sample dataset
from pycaret.datasets import get_data
import pandas as pd

data = pd.read_csv('data\Kangaroo_cleaned.csv')
```

```python
# Functionnal API
from pycaret.regression import RegressionExperiment
s = RegressionExperiment()
s.setup(data, target = 'price', session_id = 123)
```

```python
best = s.compare_models()
print(best)
```

# Analyze Model

```python
s.evaluate_model(best)
```

```python
s.plot_model(best, plot = 'residuals')
```

```python
s.plot_model(best, plot = 'learning')
```

```python
s.plot_model(best, plot= 'error')
```

# Predictions

```python
s.predict_model(best)
```

# Save model

```python
s.save_model(best, 'Light Gradient Boosting Machine')
```
